```python
import pandas as pd
import numpy as np
```

```python
df2013 = pd.read_csv('./data/standings/2013.csv')
df2014 = pd.read_csv('./data/standings/2014.csv')

df2013 = df2013.assign(NextPoints= [0] * 20)
df2014 = df2014.assign(NextPoints= [0] * 20)

for row in df2013.itertuples():
    a= df2014.loc[df2014['Team'] == row.Team, 'Points'].iloc[0]
    df2013.at[row.Index, 'NextPoints'] = a
```
